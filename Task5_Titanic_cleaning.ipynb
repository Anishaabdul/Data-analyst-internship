```python
df.drop_duplicates(inplace=True)
```

```python
df['Survived'] = df['Survived'].astype('category')
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

## Data Cleaning Summary

1. Loaded the Titanic dataset using pandas.
2. Inspected the dataset using head() and info().
3. Identified missing values using isnull().
4. Filled missing Age values with the median.
5. Filled missing Embarked values with the mode.
6. Dropped the Cabin column due to excessive missing values.
7. Removed duplicate records from the dataset.
8. Converted the Survived column to categorical datatype.
9. Created a new Age_Group feature for analysis.
10. Saved the cleaned dataset as cleaned_data.csv.
